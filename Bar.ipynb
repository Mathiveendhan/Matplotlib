```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
age_x=[22,23,24,25,26,27,28,29,30]
salary_y=[2000,3000,4000,5000,6000,7000,8000,9000,10000]
experience_z=[30,31,32,33,34,35,36,37,38]
```

```python
plt.title('salary_procedure')
plt.xlabel('age')
plt.ylabel('salary')
```

```python
plt.bar(age_x,salary_y)
```

```python
plt.bar(age_x,salary_y)
plt.title('salary_procedure')
plt.xlabel('age')
plt.ylabel('salary')
```

```python
plt.bar(age_x,experience_z)
```

```python
plt.bar(age_x,salary_y)
plt.bar(experience_z,age_x)
```

```python
plt.bar(age_x,salary_y)
plt.bar(experience_z,salary_y)
```

```python
plt.bar(age_x,salary_y,color='b')
plt.bar(experience_z,salary_y,color='g')
```

```python
a=[2304,2305,2306,2307,2308,2309,2400,2401,2402]
b=[10001,880,803,994,600,1007,10007,1000,1943]
```

```python
x_indexes=np.arange(len(age_x))
width=0.25
plt.bar(x_indexes+width,salary_y,width=width)
plt.bar(x_indexes,a,width=width)
plt.bar(x_indexes-width,b,width=width)
plt.title('salary_procedure')
plt.xlabel('age')
plt.ylabel('salary')
```
