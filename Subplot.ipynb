```python
import pandas as pd
import matplotlib.pyplot as plt

heart_data = pd.read_csv(r"C:\Users\Admin\OneDrive\Documents\Dataset\heart_data.csv")

# Create a figure and a grid of subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(7, 6))


axs[0, 0].scatter(heart_data['age'], heart_data['max_hr'])
axs[0, 0].set_title('Age vs. Max Heart Rate')

axs[0, 1].scatter(heart_data['cholesterol'], heart_data['st_depression'])
axs[0, 1].set_title('Cholesterol vs. ST Depression')

axs[1, 0].scatter(heart_data['rest_bp'], heart_data['cholesterol'])
axs[1, 0].set_title('Resting Blood Pressure vs. Cholesterol')

axs[1, 1].scatter(heart_data['age'], heart_data['cholesterol'])
axs[1, 1].set_title('Age vs. Cholesterol')
fig.suptitle('Relationships in Heart Data')

plt.tight_layout()
plt.show()
```
