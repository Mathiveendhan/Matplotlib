```python
import matplotlib.pyplot as plt
```

```python
x=[1,1,2,3,4,2,3,2,4,4,5]
y=[1,2,2,2,3,4,3,5,4,1,1]
z=[1,1,1,2,2,3,3,3,4,5,6]
l=[1,2,1,3,4,5,2,1,3,3,4]
```

```python
labels=['X','Y','Z','L']
plt.stackplot(x,y,z,l,labels=labels)
plt.legend(loc='upper left')
plt.show()
```

```python
X=[1,2,3,4,5,6,7,8,9]

a=[1,2,3,3,4,4,4,4,5]
b=[1,1,1,1,2,2,2,3,4]
c=[1,1,1,2,2,2,3,3,3]
```

```python
plt.style.use("fivethirtyeight")
```

```python
labels=['X','a','b','c']
plt.stackplot(X,a,b,c,labels=labels)
plt.legend(loc='upper right')
plt.show()
```

```python
labels=['X','a','b','c']
colors=['m','c','r','k']
plt.stackplot(X,a,b,c,labels=labels,colors=colors)
plt.legend(loc='upper right')
plt.show()
```

```python
labels=['X','a','b','c']
colors=['m','c','r','k']
plt.stackplot(X,a,b,c,labels=labels,colors=colors)
plt.legend(loc=(0.5,0.9))
plt.show()
```
